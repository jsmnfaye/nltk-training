```
# split sentence
words = sentence.split()
print(words)
```

```
# loop three iterations
for count in range(3):
  n = count + 1
  current_grams = ngrams(words, n)    # save current grams
  if n==1:
    print('Unigrams: ')
  elif n==2:
    print('\nBigrams: ')
  elif n==3:
    print('\nTrigrams: ')
    
  # print the n-grams of current n-gram
  for g in current_grams:
    print(g, end='')
```
